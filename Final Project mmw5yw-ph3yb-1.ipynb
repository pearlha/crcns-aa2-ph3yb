## **Introduction**

Our project is interested in the neural computations associated with zebra finches as their neurons respond to different types of sound stimuli that range between natural and unnatural sounds. 

The overarching goal of the project is to assess the finchesâ€™ optimization or preference for a certain song type by evaluating neuron spikes in their song control system. The song control system of these birds receive auditory input from ascending, primary auditory pathways in the cochlear nuclei that project into the Nucleus Oviodalis (OV), which then eventually projects the signals into the forebrain. 

Therefore, we understand that the OV is an important brain nucleus in song learning in zebra finches, but we are unsure of its ability in determining preference between different types of song stimuli.

## General Layout of Experiment

The birds were tested for multiple anatomical brain regions:
- Nucleus Oviodalis (OV)
- Caudal Mesopallium (CM)
- Primary Forebrain (from subareas L1, L2a, L2b, L3 and L)
- Mesencephalicus lateral dorsalis (MLd)

Each bird is tested with stimuli from two or all three different classes of sounds.

The spiking respone was recorded for the stimuli.

## **Classes of sound**

The sounds used to stimulate the Zebra Finches ranged from least natural, medium, most natural.

These are listed below in order of most to least natural:

- **Conspecific:** natural birdsong from Zebra Finches
- **Songrip:** ripple stimuli with same modulation of song
- **Flatrip:** modulation limited noise with flat modulation spectrum

The Flatrip sound has the most white noise and is the most unnatural to the birds.

## Hypothesis

Our objective is to count the number of spikes for each class to come up with a general average for each song type. 

Our **hypothesis** is that the conspecific songs will produce the largest amount of spikes, indicating that the finches have an affinity for natural sounds.

#### Next

We begin using a nested for loop to sort through the conspecific vs. flatrip for each bird. 

We accessed all of the spikes for each bird and averaged them. We then added this average to a list.

Conspec_list and flatrip_list gives us a list of the averaged spike trials of the total number of spikes for a stimuli for each bird.

We then plotted each list against the number of birds.

**Note:** the number of birds averaged is different for conspecific and flatrip because one of the birds did not have spike data for the flatrip sound.

**Note:** for accessing files downloaded on a different computer, insert an individualized directory for " addr = "

This line is essential for accessing the files downloaded on an individual's computer.

**ALSO NOTE:** > '+ i +' < must be inserted in place of the bird name (Ex: obla1305_7) so the for loop can cycle through. (Purely between the ><)

All that must be essentially changed is the directory path up to the downloaded crcns folder --> /aa2/all_cells/'+ i +'/conspecific' can be THE SAME

```python
# First for loop for conspecific

conspec_list = []

# addr = must be changed to the specific directory of the location of the user's download data

for i in OV_birds_conspec:
    addr = '/Users/madelinewilliams/Downloads/crcns 2/aa2/all_cells/'+ i +'/conspecific'
    os.chdir(addr)
    conspec_total = 0
    for stim in spike_num_conspec:
        # count number of spikes
        stim_spike = open(stim,"r")
        contents = stim_spike.read()
        spike_times = len(contents.split())

        #finding the mean number of spikes by adding spike times to an empty list
        conspec_total = conspec_total + spike_times
    # calculating the average
    conspec_avg = conspec_total / 20
    
    # adding averages to an empty list for plotting
    conspec_list.append(conspec_avg)

# Creating an index to use for the x axis so each bird name is not shown on the plot (becomes messy)
index_list_conspec = []
for i in OV_birds_conspec:
    index_num = OV_birds_conspec.index(i)
    index_list_conspec.append(index_num)

# printing the conspec_list to show the averages
print(conspec_list)
print('The number of conspecific averages for the birds is', len(conspec_list))


plt.plot(index_list_conspec, conspec_list)
plt.title(r'Conspecific Averages')
plt.xlabel('Birds')
plt.ylabel('Average Spike')
plt.show()
```

```python
# Next loop for Flatrip
# no flatrip for pupi0414_2

flatrip_list = []

# addr = must be changed to the specific directory of the location of the user's download data

for i in OV_birds_flatrip:
    addr = '/Users/madelinewilliams/Downloads/crcns 2/aa2/all_cells/'+ i +'/conspecific'
    os.chdir(addr)
    flatrip_total = 0
    for stim in spike_num_flatrip:
        stim_spike = open(stim, "r")
        contents = stim_spike.read()
        spike_times = len(contents.split())
        flatrip_total = flatrip_total + spike_times
    flatrip_avg = flatrip_total / 10
    flatrip_list.append(flatrip_avg)

# Creating an index to use for the x axis so each bird name is not shown on the plot (becomes messy)
index_list_flatrip = []
for i in OV_birds_flatrip:
    index_num = OV_birds_flatrip.index(i)
    index_list_flatrip.append(index_num)
    
# printing the flatrip_list to show the averages
print(flatrip_list)
```

**Quick note:** here we are calculating the total averages of total number of spikes for the birds for the conspecific category and the flatrip category.

We have done so, as you will see, because below when we attempt to graph the flatrip spikes, there is an absence of data for a bird that we input at None to graph correctly.

We calculate for mean averages here because a None type data in a sum creates error.

```python
print('The number of original flatrip averages for the birds is', len(flatrip_list))


conspec_final_avg = sum(conspec_list) / len(conspec_list)

flatrip_final_avg = sum(flatrip_list) / len(flatrip_list)

print('The average of all conspecific spikes is', conspec_final_avg)

print('The average of all flatrip spikes is', flatrip_final_avg)
```
